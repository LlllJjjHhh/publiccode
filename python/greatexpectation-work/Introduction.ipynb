- Remove the productivity costs and operational risk

## What is Great Expectations?
- A flexible, declarative syntax for describing the expected shape of data
- Data Pipeline Testing tools at development and production levels
- In Exploration phase, documenting latent knowledge about the shape, format, and content of data
- In production phase, Expectations are a powerful tool for testing. 
- Enabling teams to quickly zero in on potential errors and determine how to respond
